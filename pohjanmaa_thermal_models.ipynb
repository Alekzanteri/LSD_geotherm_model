## Model 1

```python
T_m=500
A=0.72
L=50
v=-1
k=25
model1=thermal_model(name='model2', depth=L)
model1.calculate_geotherm(T_m, A, L, k, v)
model1.collide(v=1.74, thickening=16, time=7, step=None)
model1.collide(v=1.85, thickening=16, time=7, step=None)
model1.collide(v=2, thickening=16, time=7, step=None)
model1.collide(v=2.15, thickening=16, time=7, step=None)
model1.collide(v=2.28, thickening=16, time=7, step=None)
model1.plot_models(ylim=70)
plt.savefig('figures/model_1.png')
```
